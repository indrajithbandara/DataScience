```python
# 75% of the data is under 17.00%
# means comparatively low CPU utilization by the system

# lets plot the scatter plot matrix between the features
from pandas.tools.plotting import scatter_matrix

features = sysActivity[sysActivity.columns[1:-3]]
# correlation matrix 
correlationMat = features.corr()
correlationMat
```

```python
highlyCorrelated = {}

for row in correlationMat:
    count = 0
    highlyCorrelated.setdefault(row, [])
    for val in correlationMat[row]:
        if (features.columns[count] != row) and (abs(val) > 0.7):
            highlyCorrelated[row].append(features.columns[count])
        count += 1
```

```python
# lets model the relationship of free disk swap block with % CPU utilization (user)
scaledFeature = (sysActivity.freeswap - sysActivity.freeswap.mean()) / (sysActivity.freeswap.std())
```

```python
plt.hist(scaledFeature)
```

```python
scaledFeatureReshaped = scaledFeature.values.reshape(scaledFeature.shape[0], 1)
```

```python
print scaledFeatureReshaped.shape
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaledFeatureReshaped, sysActivity.usr, test_size=0.2)
```

```python
print X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# display the training set
plt.scatter(X_train, y_train, alpha=0.2)
plt.xlabel('Free Disk Swap Block')
plt.ylabel('% CPU utilization(user)')
plt.title('Training Set')
```

```python
# display the test set
plt.scatter(X_test, y_test, alpha=0.2)
plt.xlabel('Free Disk Swap Block')
plt.ylabel('% CPU utilization(user)')
plt.title('Training Set')
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
print (regr.coef_)
print (regr.intercept_)
```

```python
# The mean square error
print("Training error: ", np.mean((regr.predict(X_train) - y_train) ** 2))
print("Test     error: ", np.mean((regr.predict(X_test) - y_test) ** 2))
```

```python
# Plotting data and linear model

# Visualises dots, where each dot represent a data exaple and corresponding teacher
plt.scatter(X_train, y_train,  color='black', alpha=0.2)
# Plots the linear model
plt.plot(X_train, regr.predict(X_train), color='blue', linewidth=3);
plt.xlabel('Data')
plt.ylabel('Target')
```

```python
# Visualises dots, where each dot represent a data exaple and corresponding teacher
plt.scatter(X_test, y_test,  color='black', alpha=0.2)
# Plots the linear model
plt.plot(X_test, regr.predict(X_test), color='blue', linewidth=3);
plt.xlabel('Data')
plt.ylabel('Target');
```

```python
predicted_train = regr.predict(X_train)
predicted_test = regr.predict(X_test)
```

```python
# Lets look at the residuals, on both the sets
plt.scatter(predicted_train, predicted_train - y_train, c='b', s=40, alpha=0.2)
plt.scatter(predicted_test, predicted_test - y_test, c='g', s=40, alpha=0.2)
```

```python
# lets print out the R-squared value
print regr.score(X_train, y_train), regr.score(X_test, y_test)
```
