Problem Description
===================

Predict a computer system activity from system performance measures.The computerActivity database records various performance measures, such as the bytes read/written from sytem memory, from a Sun Sparctation 20/712 with 2 cpus and 128 Mbyte of main memory.

For detail <a href='http://www.cs.toronto.edu/~delve/data/comp-activ/compActivDetail.html'> info </a>

```python
columnNames = ['time', 'lread', 'lwrite', 'scall', 'sread', 'swrite', 'fork', 'exec', 'rchar', 'wchar', 'pgout', 'ppgout',
               'pgfree', 'pgscan', 'atch', 'pgin', 'ppgin', 'pflt', 'vflt', 'runqsz', 'runocc', 'freemem',
               'freeswap', 'usr', 'sys', 'wio', 'idle']
```

```python
def loadData(location):
    sysActivity = pd.read_table(location, sep=' ', header=None, usecols=np.arange(0, 27), names=columnNames)
    return sysActivity
```

```python
# another take on the analysis
cpuActivity_dat = loadData('./data/comp-activ/comp-activ/Dataset.data/Dataset.data'); cpuActivity_dat.head()
```

```python
# lets produce the statistical summary for the dataset
cpuActivity_dat.describe()
```

```python
# looks like all the feature are on very different scale
def scaleFeatures(df, features):
    df = (df[features] - df[features].mean()) / (df[features].std())
    return df
```

```python
scaledCpuActivity_dat = scaleFeatures(cpuActivity_dat, cpuActivity_dat.columns[1:-4])
```

```python
# seems like there are a few corelated variables with
# our response varibale usr

# lets consider two of them
# 1. freeswap - Free Disk Swap Blocks
# 2. runocc - % time that run queue is occupied
```

```python
cpuActivity_dat.freeswap.describe()
```

```python
# large standard deviation
# large deviation between min and max

# lets visualize
cpuActivity_dat.freeswap.plot(kind='hist')
plt.xlabel('Free Disk Swap Blocks')
```

```python
# lets perform log transformation on this variable
freeswapTransformed = np.log10(cpuActivity_dat.freeswap + 1)
```

```python
plt.hist(freeswapTransformed, bins=50)
plt.xlabel('Log(Free Disk Swap Blocks)')
```

```python
# as we can see there are two different
freeswapThreshold = 10 ** 1.7
freeswapBelowThreshold = cpuActivity_dat[cpuActivity_dat.freeswap < freeswapThreshold].freeswap
cpuUsrBelowThreshold = cpuActivity_dat[cpuActivity_dat.freeswap < freeswapThreshold].usr

plt.scatter(freeswapBelowThreshold, cpuUsrBelowThreshold, alpha=0.08)
plt.title('Relationship between number of free disk swap blocks and cpu utilization in usr mode')
plt.xlabel('Number of free disk swap blocks below threshold')
plt.ylabel('Cpu utilization in user mode')
```

```python
freeswapAboveThreshold = cpuActivity_dat[cpuActivity_dat.freeswap > freeswapThreshold].freeswap
cpuUsrAboveThreshold = cpuActivity_dat[cpuActivity_dat.freeswap > freeswapThreshold].usr

plt.scatter(freeswapAboveThreshold, cpuUsrAboveThreshold, alpha=0.08)
plt.title('Relationship between number of free disk swap blocks and cpu utilization in usr mode')
plt.xlabel('Number of free disk swap blocks above threshold')
plt.ylabel('Cpu utilization in user mode')
```

```python
# so for values of free disk swap blocks below threshold
# cpu utilization in user mode is very low
# and for others cpu utilization in user mode is greater than
# 50 %

# Let's create the model

scaledCpuActivity_dat['usr'] = cpuActivity_dat.usr
```

```python
est_s = smf.ols(formula='usr ~ freeswap', data=scaledCpuActivity_dat).fit()
```

```python
# View the model estimates
est_s.summary()
```

```python
freeswapFrame = pd.DataFrame({'freeswap': scaledCpuActivity_dat.freeswap.values})
y_hat = est_s.predict(freeswapFrame)
```

```python
plt.scatter(scaledCpuActivity_dat.freeswap, scaledCpuActivity_dat.usr, alpha=.08)
plt.plot(scaledCpuActivity_dat.freeswap, y_hat, 'r', linewidth=2, alpha=.9)
```

```python
# Lets plot the residuals
plt.scatter(est_s.predict(freeswapFrame), est_s.resid, alpha=0.3)
plt.xlabel('Predicted Cpu utilization in user mode')
plt.ylabel('Residuals')
```

```python
# lets print out the rmse of the prediction
print 'Root Mean Square Error ', sqrt(np.sum((y_hat - scaledCpuActivity_dat.usr.values) ** 2) / (scaledCpuActivity_dat.usr.shape[0] - 1))
```

```python
# Lets see how the second degree ploynomial would work
poly_2 = smf.ols(formula='usr ~ freeswap + I(freeswap ** 2.0)', data=scaledCpuActivity_dat).fit()
poly_2.summary()
```

```python
plt.scatter(scaledCpuActivity_dat.freeswap, scaledCpuActivity_dat.usr, alpha=0.08)
plt.xlabel('Number of free disk swap blocks')
plt.ylabel('Cpu utilization in user mode')
plt.plot(freeswapFrame.freeswap, poly_2.predict(freeswapFrame), 'g-', label='Poly n=2 $R^2$=%.2f' % poly_2.rsquared, 
         alpha=0.9)

plt.legend()
```

```python
# Lets plot the residuals
plt.scatter(poly_2.predict(freeswapFrame), poly_2.resid, alpha=0.3)
plt.xlabel('Predicted Cpu utilization in user mode')
plt.ylabel('Residuals')
```

```python
print 'Root Mean Square Error is ', sqrt(np.sum((poly_2.predict(freeswapFrame) - scaledCpuActivity_dat.usr.values) ** 2) / (scaledCpuActivity_dat.usr.shape[0] - 2))
```

```python
# lets see what happens if we add the interaction between
# number of free disk space and free mem pages

interaction = smf.ols(formula='usr ~ freeswap + freeswap:freemem', data=scaledCpuActivity_dat).fit()
interaction.summary()
```

```python
interactionDataFrame = pd.DataFrame({'freeswap': scaledCpuActivity_dat.freeswap.values, 'freemem': scaledCpuActivity_dat.freemem.values})
plt.scatter(scaledCpuActivity_dat.freeswap, scaledCpuActivity_dat.usr, alpha=0.08)
plt.xlabel('Number of free disk swap blocks')
plt.ylabel('Cpu utilization in user mode')
plt.plot(interactionDataFrame.freeswap, interaction.predict(interactionDataFrame), 'g-', label='Interaction $R^2$=%.2f' % interaction.rsquared, 
         alpha=0.9)

plt.legend()
```

```python
# Lets plot the residuals
plt.scatter(interaction.predict(interactionDataFrame), interaction.resid, alpha=0.3)
plt.xlabel('Predicted Cpu utilization in user mode')
plt.ylabel('Residuals')
```

```python
print 'Root Mean Square Error ', sqrt(np.sum( (interaction.predict(interactionDataFrame) - scaledCpuActivity_dat.usr) ** 2 )/ (scaledCpuActivity_dat.usr.shape[0] - 2))
```

```python
# lets take the 2 degree polynomial
# and validate the model

from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaledCpuActivity_dat.freeswap, scaledCpuActivity_dat.usr, test_size=0.2)
```

```python
train_df = pd.DataFrame({'freeswap': X_train, 'usr': y_train})
poly_2_val = smf.ols(formula='usr ~ freeswap + I(freeswap ** 2.0)', data=train_df).fit()
poly_2_val.summary()
```

```python
test_df = pd.DataFrame({'freeswap': X_test})
plt.scatter(test_df.freeswap, y_test, alpha=0.08)
plt.xlabel('Number of free disk swap blocks')
plt.ylabel('Cpu utilization in user mode')
plt.plot(test_df.freeswap, poly_2_val.predict(test_df), 'g-', label='Poly n=2 $R^2$=%.2f' % poly_2_val.rsquared, 
         alpha=0.9)

plt.legend()
```
