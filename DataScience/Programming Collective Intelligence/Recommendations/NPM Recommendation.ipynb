<b>Tag Similarity</b>
<p>npm is a package manager for javascript, and is default for Node.js. As of Node.js version 0.6.3, npm is bundled and installed automatically with the environment. npm runs through the command line and manages dependencies for an application. It also allows users to install Node.js applications that are available on the <a href='https://registry.npmjs.org/'>npm registry</a>.So, I decided to fetch different modules from the npm registry with their tags. Following is the analysis to find out 5 most similar tags based on kind of modules stored for each of the tags.</p>

```python
# this is the url structure to fetch all the tags with all of their modules.
npm_registry = requests.get('http://registry.npmjs.org/-/_view/byKeyword?startkey=%5B%22a%22%5D&endkey=%5B%22b%22,%7B%7D%5D&group_level=2').text
```

```python
# convert it into JSON representation
npm_registry_json = json.loads(npm_registry)
```

```python
# present a sample entry
# explain the number

print len(npm_registry_json['rows'])
```

```python
# a row of this data which contain our data in 'key', first element of this
# array is the name of the tag and the second element is the module name
# only a single module name is mentioned per row
# so we have to merge all the modules for a tag in a single dict

print npm_registry_json['rows'][0]
```

```python
# so there are 22,139 rows in the dataset which contain multiple entries
# for a single tag


def initializeTagDict(data):
    '''
    Initialize an empty for dictionary for each of the tags
    which will be populate later with the modules
    '''
    
    tag_dict = {}
    
    for row in data['rows']:
        tag = row['key'][0]
        tag_dict[tag] = {}
    
    return tag_dict

def fillTags(tag_dict, data):
    '''
    Fill each of the tags with the modules associated with the tag
    e.g. {tag_1: { module_1: 1.0, module_2: 1.0 ...}, {tag_2: { module_1: 1.0, module_3: 1.0}, ...}}
    '''
    for row in data['rows']:
        tag = row['key'][0]
        module = row['key'][1]
        tag_dict[tag][module] = 1.0
    
    return tag_dict
```

```python
npm_tags = initializeTagDict(npm_registry_json)
```

```python
# now after merging all the rows
# we are left with 2398 unique tags

print len(npm_tags.keys())
```

```python
tag_dict = fillTags(npm_tags, npm_registry_json)
```

Similarity Function
-------------------

<p>To measure how similar two items are we have to use similarity functions like Euclidean, Pearson Correlation, cosine similairty etc. but for this problem I went ahead with <a href='http://en.wikipedia.org/wiki/Jaccard_index'>Jaccard Index </a> (Tanimoto score). It is useful in cases where we use 1.0 for including something in the set and 0.0 for omitting instead of a using proper ratings system. Well I will try various other similarity functions and see how they behave.
</p>

```python
def sim_tanimoto(prefs, p1, p2):
    # Get the list of mutually rated items
    si = {}
    
    for item in prefs[p1]:
        if item in prefs[p2]: si[item] = 1
    
    # Find the number of elements
    n = len(si)
    
    # if there are no ratings in common, return 0
    if n == 0: return 0
    
    intersection = prefs[p1].viewkeys() & prefs[p2].viewkeys()
    union = prefs[p1].viewkeys() | prefs[p1].viewkeys()
        
    len_intersection = len(intersection) * 1.0
    len_union = len(union) * 1.0
  
    r = len_intersection / len_union
    
    return r

# Returns the best matches for person from the prefs dictionary.
# Number of results and similarity function are optional params.

def topMatches(tags_dict, tag, n=5, similarity=sim_tanimoto):
    scores = [(similarity(tags_dict, tag, other), other) for other in tags_dict if other != tag]
    
    # scores the list so the highest scores appear at the top
    scores.sort()
    scores.reverse()
    return scores[0:n]

def calculateSimilarTags(tags_dict, n=10):
    # Create a dictionary of items showing which other
    # items they are most similar to.
    
    result = {}
    c = 0
    for tag in tags_dict:
        # status update for large dataset
        c += 1
        if c % 100 == 0: print '%d / %d' %(c, len(tag_dict))
        # Find the most similar items to this one
        scores = topMatches(tags_dict, tag, n=n, similarity=sim_tanimoto)
        result[tag] = scores
    
    return result
```

```python
# tags similarity dictionary which contains a list of top 5 tags with their similarity scores for a given tag.
tags_sim = calculateSimilarTags(tag_dict, 5)
```
