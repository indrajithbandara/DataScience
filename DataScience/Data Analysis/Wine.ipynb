Description about the dataset
-----------------------------

<p>These data are the results of a chemical analysis of wines grown in the same region in Italy but derived from three different cultivars. The analysis determined the quantities of 13 constituents found in each of the three types of wines.</p>

<b><u>Attributes</u></b>
<ul>
    <li>
        Alcohol: percentage of alcohol 
    </li>
    <li>
        Malic Acid: organic compound made by all living organisms, used as food additive, for more <a href='http://en.wikipedia.org/wiki/Malic_acid'>info</a>
    </li>
    <li>
        Ash: One of the important indicators in wine quality determination.
    </li>
    <li>
        Alkalinity of ash: How much alkaline ash is? 
    </li>
    <li>
        Magnesium: Magensium content in a wine sample
    </li>
    <li>
        Total phenols: The phenolic content in wine refers to the phenolic compounds—natural phenol and polyphenols—in wine, which include a large group of several hundred chemical compounds that affect the taste, color and mouthfeel of wine.
    </li>
    <li>
        Flavonoids: Flavonoids are the most important plant pigments for flower coloration, producing yellow or red/blue pigmentation in petals designed to attract pollinator animals
    </li>
    <li>
        Nonflavanoid phenols
    </li>
    <li>
        Proanthocyanins: Proanthocyanidins are a class of polyphenols found in a variety of plants. Chemically, they are oligomeric flavonoids.
    </li>
    <li>
        Color intensity
    </li>
    <li>
        Hue
    </li>
    <li>
        OD280/OD315 of diluted wines
    </li>
    <li>
        Proline : amino acid
    </li>
</ul>

```python
# now column names are defined
# lets just re-import it with column names defined
column_names = ['wine_type', 'alcohol', 'malic_acid', 'ash', 'alkalinity_of_ash', 'magnesium', 
                'total phenols', 'flavanoids', 'nonflavanoids_phenols', 'proanthocyanins',
                'color_intensity', 'hue', 'OD280/OD315_of_diluted_wines', 'proline']
```

```python
# lets see the summary statistic for the dataframes
wine[wine.columns[1:]].describe()
```

<p>So all of the attributes are on different scale.</p>

```python
# lets plot all the distributions on different scales
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(15, 15))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

row = col = 0
for i, var in enumerate(wine.columns[1:]):
    wine[var].plot(ax=axes[row][col], kind='hist', bins=20, title=var, sharex=False,
                   color='k', alpha=0.5)
    if col == 2:
        row += 1
    col = (col + 1) % 3
```

<b>Malic acid, Magnesium and Proline have right skewed distribution. Nonflavonoids phenols have abnormally low frequency around 0.35, this might require some investigation.</b>

```python
# lets plot the box plot of individual to detect any outliers in the data.
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(15, 20))
plt.subplots_adjust(wspace=0.5, hspace=0.5)
color = dict(boxes='DarkGreen', whiskers='DarkOrange',
             medians='DarkBlue', caps='Gray')

row = col = 0
for i, var in enumerate(wine.columns[1:]):
    wine[var].plot(ax=axes[row][col], kind='box', title=var, sharex=False,
                   color=color, sym='r+')
    if col == 2:
        row += 1
    col = (col + 1) % 3
```

<b>Malic Acid, Magnesium and color intensity attributes have some outliers, lets go into detail.</b>

```python
wine.malic_acid.describe()
```

<b>75 percent of the values are less than 3.0</b>

```python
wine[wine.malic_acid > 4.0]
```

```python
wine[wine.malic_acid > 4.0].wine_type.value_counts()
```

<b>Most of the wines with high content of malic acid either belong to class 2 or class 3.</b>

```python
# Lets plot the box plots for all of the attributes group by the wine type
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(10, 20))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

row = col = 0
for i, var in enumerate(wine.columns[1:]):
    wine[['wine_type', var]].boxplot(by='wine_type', ax=axes[row][col])
    if col == 2:
        row += 1
    col = (col + 1) % 3
```

<b>This graph shows that all the attributes have some distinguishing power, e.g. Proline value for class 1 wine is significantly higher than other than wines of other classes.</b>

```python
# Lets plot to see whether there is any new correlation in the attributes or not.
wine[wine.columns[1:]].corr()
```

<b>So there are some correlated variables such as OD280/OD315 of diluted wines and flavanoids with strong positive correlation of 0.787, similarly total_phenols is also strongly correlated with OD280/OD315 of diluted wines. Proline and alcohol are also correlated.</b>

```python
# first just scale all the attributes
wine[wine.columns[1:]] = (wine[wine.columns[1:]] - wine[wine.columns[1:]].mean()) / (wine[wine.columns[1:]].std())
```

```python
wine.head()
```

```python
from sklearn.lda import LDA
from sklearn.cross_validation import train_test_split
```

```python
X = wine[wine.columns[1:]]
Y = wine.wine_type
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
print X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
lda = LDA()
clf = lda.fit(X_train, y_train)
print clf.score(X_train, y_train)
```

```python
print clf.score(X_test, y_test)
```

```python
# Quadratic Discriminant Analysis
from sklearn.qda import QDA
```

```python
qda = QDA()
clf2 = qda.fit(X_train, y_train)
print clf2.score(X_train, y_train)
```

```python
print clf2.score(X_test, y_test)
```
