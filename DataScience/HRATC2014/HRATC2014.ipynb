```python
#lets see the correlation matrix for the attributes
ground_truth[ground_truth.columns[1:]].corr()
```

```python
from pandas.tools.plotting import scatter_matrix
```

```python
scatter_matrix(ground_truth[ground_truth.columns[1:]], alpha=0.2, figsize=(10, 10), diagonal='kde')
```

<b>Altitude has right skewed distribution, which can be investigated further.</b>

```python
ground_truth.Type.unique()
color_dict = {'Surrogate Mine': 'red', 'Coke Can': 'blue', 'Small Metal Piece': 'black',
       'Large Metal Piece': 'yellow', '3 Metal Screws': 'orange'}
color_map = ground_truth.Type.map(lambda x: color_dict[x])
```

```python
# lets plot the longitude vs latitude using altitude as the third dimension
plt.scatter(ground_truth.Latitude, ground_truth.Longitude, c=color_map)
plt.xlim(40.184709, 40.184742)
```

```python
# statistical summary
ground_truth[ground_truth.columns[1:]].describe()
```

<p>Standard deviation for all the three attributes are near 0, which suggests that no variation in lat, long which means all the data originated from a single source.</p>
